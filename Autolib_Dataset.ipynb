```python
# Importing the necessary libraries for pandas and numpy

import pandas as pd
import numpy as np

```

```python
# Loading dataset

df = pd.read_csv('http://bit.ly/autolib_dataset')

df
```

```python
#We then explore our data

df.info()
```

```python
# We then go to Data Cleaning and begin with the column names
# Changing column names to lowercase

df.columns = map(str.lower, df.columns)

columns = ['address', 'cars', 'bluecar_counter', 'utilib_counter', 'utilib_1.4_counter', 'charge_slots', 'charging_status', 'city', 'displayed_comment', 'id', 'kind', 'geo_point', 'postal_code', 'public_name', 'rental_status', 'scheduled_at', 'slots', 'station_type', 'status', 'subscription_status', 'year', 'month', 'day', 'hour', 'minute']

df.columns = columns

df.columns

```

```python
# Checking the shape of the new dataset

df.shape
```

```python
# Previewing the data

df.head()
```

```python
# Check for null values

df.isnull().sum()
```

```python
# Drop the displayed_comment column and scheduled_at, too many missing columns

df.drop('displayed_comment', axis= 1, inplace= True)
```

```python
df.drop('scheduled_at', axis= 1, inplace= True)
```

```python
#Check for null values again

df.isnull().sum()
```

```python
#Check data type of dataset again

df.info()
```

```python
#Convert year, month and day column to date column

df['date'] = df['year'].map(str) + '-' + df['month'].map(str) + '-' + df['day'].map(str)
```

```python
#Combining hour and minute column to time column

df['time'] = df['hour'].map(str) + df['minute'].map(str) 
```

```python
#Changing date and time columns to datetime types

df.date = pd.to_datetime(df.date)
```

```python
df['time'] = (df['time'].apply(lambda x: pd.to_datetime(x, format = '%H%M')).dt.strftime('%H:%M'))
```
